## Lab 3 - Bài 1

```python
# Code lab 3 bài 1 ở đây
# khai báo dãy số ban đầu chưa có phần tử
day_so = []
# khởi tạo biến
tong = 0
dem = 0
# Nhập dãy số
while True:
    so = input("Nhập số: ")
    if so.lower() == "q":
        print("Thoát chương trình")
        break
    else:
        try:
            phan_tu = int(so)
            day_so.append(phan_tu) # phần try sẽ sử lý những phần đúng
        except ValueError:
            print("Nhập không đúng, hãy nhập số hoặc q để thoát chương trình") # phần except sẽ sử lý những phần lỗi

# sắp xếp dãy số tăng dần
day_so.sort()
print("Dãy số đã sắp xếp tăng dần")
print(day_so)
print("Giá trị nhỏ nhất của dãy số:", day_so[0])
print("Giá trị lớn nhất của dãy số:", day_so[-1])
print("")

# sắp xếp dãy số giảm dần
day_so.sort(reverse = True)
print("Dãy số đã sắp xếp giảm dần")
print(day_so)
print("Giá trị lớn nhất của dãy số:", day_so[0])
print("Giá trị nhỏ nhất của dãy số:", day_so[-1])


# duyệt qua danh sách dãy số Tổng và đếm các phần tử chia hết cho 3
for phan_tu in day_so:
    if phan_tu % 3 == 0:
        tong += phan_tu
        dem += 1
print("")

# xuất thông tin
if dem != 0:
    print("tổng giá trị cần tìm", tong/dem)
else:
    print("Không có phần tử trong dãy số chia hết cho 3")
```

## Lab 3 - Bài 2

```python
# Code lab 3  bài 2 ở đây
studen_name = {
    "Họ tên": "Nguyễn Văn A",
    "Điểm": 7,
    "Xếp loại": "Khá" 
}

Điểm_SV = studen_name["Điểm"]

if Điểm_SV >= 9:
    print("Xuất sắc")
elif Điểm_SV >= 8:
    print("Giỏi")
elif Điểm_SV >= 7:
    print("Khá")
elif Điểm_SV >= 5:
    print("Trung bình")
else:
    print("Yếu")
```

```python
list_student = []
so_sv = int(input("Nhập số lượng sinh viên: "))
for i in range(so_sv):
    thong_tin_SV = input("Nhập họ tên sinh viên: ")
    ma_SV = input("Nhập mã sinh viên: ")
    diem_SV = float(input("Nhập điểm sinh viên: "))

    if diem_SV >= 9:
        xep_loai = "Xuất sắc"
    elif diem_SV >= 8:
        xep_loai = "Giỏi"
    elif diem_SV >= 7:
        xep_loai = "Khá"
    elif diem_SV >= 5:
        xep_loai = "Trung bình"
    else:
        xep_loai = "Yếu"

    sinh_vien = {
        "Họ tên": thong_tin_SV,
        "Mã sinh viên": ma_SV,
        "Điểm sinh viên": diem_SV,
        "Xếp loại": xep_loai
    }

    list_student.append(sinh_vien)

print("\n====Danh sách sinh viên====")
for sinh_vien in list_student:
    print(f"Họ tên: {sinh_vien["Họ tên"]}")
    print(f"Mã sinh viên: {sinh_vien["Mã sinh viên"]}")
    print(f"Điểm sinh viên: {sinh_vien["Điểm sinh viên"]}")
    print(f"Xếp loại: {sinh_vien["Xếp loại"]}\n")

```

## Lab 3 - Bài 5: bài tập làm thêm

```python
# code lab 3  bài làm thêm ở đây 
import math
from datetime import datetime
Lich_su = []

while True:
    print("========================MENU========================")
    print("1. Phép tính cơ bản: Cộng, trừ, nhân, chia")
    print("2. Lũy thừa: x^y ")
    print("3. Căn bậc hai")
    print("4. Hàm lượng giác: Sin, cos, tan")
    print("5. Logarit: Log cơ số 10, ln, log cơ số tùy chọn")
    print("6. Giải phương trình bậc nhất")
    print("7. Giải phương trình bậc hai")
    print("8. Lịch sử: Lưu và xem lại các phép tính đã thực hiện")
    print("9. Thời gian: Hiển thị thời gian hiện tại")
    print("========================MENU========================\n")

# Khởi tạo biến    
    chọn_chương_trình = input("Nhập chọn chức năng từ 1 - 9: ")

# Thoát chương trình:
    if chọn_chương_trình == "0":
        print("Chương trình kết thúc")
        break

# Chức năng số 1:
    if chọn_chương_trình == "1":
        nhap_so_a = float(input("Nhập hệ số a: "))
        nhap_so_b = float(input("Nhập hệ số b: "))
        chọn_chức_năng = input("Chọn chức năng: +,-,*,/")

        if chọn_chức_năng == "+":
            ket_qua = nhap_so_a + nhap_so_b
            print(f"Tổng của hai số {nhap_so_a} và {nhap_so_b} = {ket_qua}\n")
            Lich_su.append(f"Phép tính cơ bản +: {nhap_so_a} + {nhap_so_b} = {ket_qua}")

        elif chọn_chức_năng == "-":
            ket_qua = nhap_so_a - nhap_so_b
            print(f"Hiệu của hai số {nhap_so_a} và {nhap_so_b} = {ket_qua}\n")
            Lich_su.append(f"Phép tính cơ bản -: {nhap_so_a} - {nhap_so_b} = {ket_qua}")

        elif chọn_chức_năng == "*":
            ket_qua = nhap_so_a * nhap_so_b
            print(f"Tích của hai số {nhap_so_a} và {nhap_so_b} = {ket_qua}\n")
            Lich_su.append(f"Phép tính cơ bản *: {nhap_so_a} * {nhap_so_b} = {ket_qua}")

        elif chọn_chức_năng == "/":
            ket_qua = nhap_so_a / nhap_so_b
            if nhap_so_b != 0:
                print(f"Thương của hai số {nhap_so_a} và {nhap_so_b} = {ket_qua}\n")
                Lich_su.append(f"Phép tính cơ bản /: {nhap_so_a} / {nhap_so_b} = {ket_qua}")
            else:
                print("Lỗi: không thể chia cho 0\n")
        else:
            print("Sai chức năng, vui lòng nhập lại\n")

# Chức năng số 2:
    elif chọn_chương_trình == "2":
        nhap_so_a = float(input("Nhập hệ số a: "))
        nhap_so_b = float(input("Nhập hệ số b: "))
        ket_qua = nhap_so_a ** nhap_so_b
        print(f"Lũy thừa: a^b là {ket_qua}\n")
        Lich_su.append(f"Lũy thừa: {nhap_so_a}^{nhap_so_b} = {ket_qua}")

# Chức năng số 3:
    elif chọn_chương_trình == "3":
        nhap_so_a = float(input("Nhập hệ số a: "))
        ket_qua = math.sqrt(nhap_so_a)
        print(f"Căn bậc hai của hệ số a là {ket_qua}\n")
        Lich_su.append(f"Căn bậc hai: của {nhap_so_a} là {ket_qua}")

# Chức năng số 4:
    elif chọn_chương_trình == "4":
        nhap_so_a = float(input("Nhập hệ số a: "))
        nhap_so_b = float(input("Nhập hệ số b: "))
        chọn_chức_năng = input("Chọn hàm lượng giác: Sin, Cos, Tan\n")
        if chọn_chức_năng == "Sin":
            ket_qua_a = math.sin(nhap_so_a)
            ket_qua_b = math.sin(nhap_so_b)
            print(f"Sin của hệ số a là {ket_qua_a} và hệ số b là {ket_qua_b}\n")
            Lich_su.append(f"Chọn hàm lượng giác: Sin của a là {ket_qua_a}, b là {ket_qua_b}")

        elif chọn_chức_năng == "Cos":
            ket_qua_a = math.cos(nhap_so_a)
            ket_qua_b = math.cos(nhap_so_b)
            print(f"Cos của hệ số a là {ket_qua_a} và hệ số b là {ket_qua_b}\n")
            Lich_su.append(f"Chọn hàm lượng giác: Cos của a là {ket_qua_a}, b là {ket_qua_b}")

        elif chọn_chức_năng == "Tan":
            ket_qua_a = math.tan(nhap_so_a)
            ket_qua_b = math.tan(nhap_so_b)
            print(f"Tan của hệ số a là {ket_qua_a} và hệ số b là {ket_qua_b}\n")
            Lich_su.append(f"Chọn hàm lượng giác: Tan của a là {ket_qua_a}, b là {ket_qua_b}")

# Chức năng số 5:
    elif chọn_chương_trình == "5":
        nhap_so_a = float(input("Nhập hệ số a: "))
        nhap_so_b = float(input("Nhập hệ số b: "))
        chọn_chức_năng = input("Chọn log10, ln, log cơ số tùy chọn: \n")
        if chọn_chức_năng == "log10":
            ket_qua_a = math.log10(nhap_so_a)
            ket_qua_b = math.log10(nhap_so_b)
            print(f"Log cơ số 10 của hệ số a là {ket_qua_a} và b là {ket_qua_b}\n")
            Lich_su.append(f"Logarit: log10 của số a là {ket_qua_a}, b là {ket_qua_b}")

        elif chọn_chức_năng == "log":
            ket_qua_a = math.log(nhap_so_a)
            ket_qua_b = math.log(nhap_so_b)
            print(f"log của hệ số a là {ket_qua_a} và b là {ket_qua_b}\n")
            Lich_su.append(f"Logarit: ln của số a là {ket_qua_a}, b là {ket_qua_b}")

        elif chọn_chức_năng == "ln":
            cơ_số_log = float(input("Nhập cơ số log: "))
            print(f"log_{cơ_số_log} ({nhap_so_a}) = {math.log(nhap_so_a, cơ_số_log)}, log_{cơ_số_log} ({nhap_so_b} = {math.log(nhap_so_b, cơ_số_log)})\n")
            Lich_su.append(f"Logarit: log cơ số tùy chọn của {nhap_so_a} cơ số {cơ_số_log} = {math.log(nhap_so_a, cơ_số_log)}, của {nhap_so_b} cơ số {cơ_số_log} = {math.log(nhap_so_a, cơ_số_log)}")

# Chức năng số 6:
    elif chọn_chương_trình == "6":
        nhap_so_a = float(input("Nhập hệ số a: "))
        nhap_so_b = float(input("Nhập hệ số b: "))
        if nhap_so_a == 0:
            if nhap_so_b == 0:
                print("Phương trình vô số nghiệm")
                Lich_su.append(f"Phương trình bậc nhất: Phương trình vô số nghiệm")
            else:
                print("Phương trình vô nghiệm")
                Lich_su.append(f"Phương trình bậc nhất: Phương trình vô nghiệm")
        else:
            print(f"Phương trình có nghiệm x = {-nhap_so_b/nhap_so_a}")
            Lich_su.append(f"Giải phương trình bậc nhất: Phương trnhf có nghiệm x = {-nhap_so_b/nhap_so_a}")

# Chức năng số 7:
    elif chọn_chương_trình == "7":
        nhap_so_a = float(input("Nhập hệ số a: "))
        nhap_so_b = float(input("Nhập hệ số b: "))
        nhap_so_c = float(input("Nhập hệ số c: "))
        if nhap_so_a == 0:
            print(f"Phương trình có nghiệm bậc nhất x = {-nhap_so_c/nhap_so_b}")
            Lich_su.append(f"Giải phương trình bậc hai: Phương trình có nghiệm bậc nhất x = {-nhap_so_c/nhap_so_b}")

        else:
            delta = nhap_so_b * nhap_so_b - 4 * nhap_so_a * nhap_so_c
            if delta < 0:
                print("Phương trình vô nghiệm")
                Lich_su.append(f"Giải phương trình bậc hai: Phương trình vô nghiệm")

            elif delta == 0:
                print(f"Phương trình có nghiệm x = {-nhap_so_b/(2*nhap_so_a)}")
                Lich_su.append(f"Giải phương trình bậc hai: Phương trình có nghiệm x = {-nhap_so_b/(2*nhap_so_a)}")

            elif delta > 0:
                print("Phương trình có 2 nghiệm phân biệt")
                print(f"X1 = {-nhap_so_b + math.sqrt(delta)/(2*nhap_so_a)}")
                print(f"X2 = {-nhap_so_b - math.sqrt(delta)/(2*nhap_so_a)}\n")
                Lich_su.append(f"Giải phương trình bậc hai:\n Phương trình có 2 nghiệm phân biệt\n X1 = {-nhap_so_b + math.sqrt(delta)/(2*nhap_so_a)}\n X2 = {-nhap_so_b - math.sqrt(delta)/(2*nhap_so_a)}\n")

# Chức năng số 8:
    elif chọn_chương_trình == "8":
        if len(Lich_su) == 0:
            print("Lịch sử trống, chưa có chức năng nào được chọn")
        else:
            print("======Lịch sử======\n")
            for i, j in enumerate(Lich_su, start = 1):
                print(f"{i}. {j}\n")

# Chức năng số 9:
    elif chọn_chương_trình == "9":
        now = datetime.now()
        format_datime = now.strftime("%d/%m/%Y %H:%M:%S")
        print(f"Thời gian hiện tại: {format_datime}\n")



```
